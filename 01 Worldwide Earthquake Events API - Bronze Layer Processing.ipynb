## Worldwide Earthquake Events API - Bronze Layer Processing

```python
import requests
import json
```

## Fetching and Saving the data as json

```python
# Construct the API URL with start and end date provided by Data Factory, formatted for geojson output.
url = f"https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime={start_date}&endtime={end_date}"

# Make the get request to fetch data
response = requests.get(url)

# Checking if the request was successful
if response.status_code == 200:
    #Get the JSON response
    data = response.json()
    data = data['features']

    # File name for data to be store in
    file_path = f'/lakehouse/default/Files/{start_date}_earthquake.json'

    # Saving the data in the specified file
    with open(file_path, 'w') as file:
        json.dump(data, file, indent=4)

    print(f"Data Successfully saved to {file_path}")

else:
    print(f"Failed to fetch data. Status Code: {response.status_code}")
```
